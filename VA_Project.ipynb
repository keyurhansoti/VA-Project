## Statewise Loan amount

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('ChoroplethData.csv')

statesname = list(data.State)

amount = list(data.LoanAmount)

index = np.arange(len(statesname))

plt.bar(index, amount)

fig_size = plt.rcParams["figure.figsize"]

fig_size[0] = 12
fig_size[1] = 9

plt.xlabel('State', fontsize=10)
plt.ylabel('Loanamount', fontsize=10)
plt.xticks(index, statesname, fontsize=8, rotation=0)
plt.title('State wise loan amount')


plt.show()
```

## Best offer based on APR 

```python
apr_data = pd.read_csv('BestOfferAPR.csv')

apr_percentage = list(apr_data.APRPercentage)

best_offer = list(apr_data.IsBestOffer)

fig_size = plt.rcParams["figure.figsize"]

fig_size[0] = 5
fig_size[1] = 9

index = np.arange(len(best_offer))

plt.bar(index, apr_percentage, color = ('#c79c00'))
plt.xlabel('Best Offer', fontsize=10)
plt.ylabel('APR', fontsize=10)
plt.xticks(index, best_offer, fontsize=10)
plt.title('Best offer based on APR')


plt.show()
```

## Total Accepted and Rejected offers

```python
lenderofferstatus_data = pd.read_csv('LenderOfferStatus.csv')

offerstatus = list(lenderofferstatus_data.OfferStatus)

statuscount = list(lenderofferstatus_data.StatusCount)

index_status = np.arange(len(offerstatus))

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 5
fig_size[1] = 9

plt.bar(index_status, statuscount, color = ('#2f4f4f'))

plt.xlabel('Offer Status', fontsize=10)
plt.ylabel('Total Offers', fontsize=10)
plt.xticks(index, offerstatus, fontsize=10)
plt.title('Total Accepted and Rejected offers')


plt.show()
```
